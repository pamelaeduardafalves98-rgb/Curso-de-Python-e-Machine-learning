# 🚀 Explorando o Python

Este notebook mostra algumas possibilidades legais do que você pode fazer com Python.

Execute cada célula e explore!

## 👋 1. Primeiro programa

```python
print("Olá, mundo! 👋")
```

## 📦 2. Variáveis e entrada de dados

```python
nome = input("Digite seu nome: ")
idade = int(input("Digite sua idade: "))
print(f"Olá {nome}, você tem {idade} anos!")
```

## 🧮 3. Calculadora simples

```python
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))
print("Soma:", a + b)
print("Multiplicação:", a * b)
```

## 🔁 4. Repetição (loop)

```python
for i in range(5):
    print("Número:", i)
```

## 🤔 5. Condições

```python
idade = int(input("Digite sua idade: "))
if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

## 🎲 6. Número aleatório

```python
import random

numero = random.randint(1, 10)
print("Número sorteado:", numero)
```

## 📋 7. Listas

```python
frutas = ["maçã", "banana", "uva"]
for fruta in frutas:
    print(fruta)
```

## 📊 8. Gráfico simples

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title("Exemplo de gráfico")
plt.show()
```

## ⚙️ 9. Automação simples

```python
with open("arquivo.txt", "w") as f:
    f.write("Python é incrível!")

print("Arquivo criado!")
```

## 🎮 10. Mini jogo (adivinhação)

```python
import random

numero = random.randint(1, 5)
palpite = int(input("Adivinhe o número (1-5): "))

if palpite == numero:
    print("Acertou! 🎉")
else:
    print(f"Errou! O número era {numero}")
```
