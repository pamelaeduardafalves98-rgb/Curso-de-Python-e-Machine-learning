# 🚀 Python Avançado com Demonstrações Visuais

Este notebook apresenta exemplos mais avançados com visualizações e aplicações práticas.

## 📊 1. Visualização de Dados

```
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,1,8,7]

plt.plot(x, y)
plt.title('Crescimento de Valores')
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.show()
```

## 🎨 2. Gráfico de dispersão

```
import matplotlib.pyplot as plt
import random

x = [random.randint(0,100) for _ in range(50)]
y = [random.randint(0,100) for _ in range(50)]

plt.scatter(x, y)
plt.title('Distribuição Aleatória')
plt.show()
```

## 🎬 3. Simulação simples (loop visual)

```
import time

for i in range(5):
    print('Carregando' + '.'*i)
    time.sleep(1)
```

## 📈 4. Análise de Dados com Pandas

```
import pandas as pd

data = {
    'Nome': ['Ana','João','Maria'],
    'Idade': [20,25,22]
}

df = pd.DataFrame(data)
print(df)
print('Média de idade:', df['Idade'].mean())
```

## 🧱 6. Programação Orientada a Objetos

```
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def apresentar(self):
        print(f'Olá, meu nome é {self.nome}')

p = Pessoa('Carlos', 30)
p.apresentar()
```

## 📂 7. Leitura de Arquivo

```
with open('arquivo.txt','w') as f:
    f.write('Linha 1\nLinha 2')

with open('arquivo.txt','r') as f:
    print(f.read())
```

## 🎲 8. Simulação (Monte Carlo simples)

```
import random

tentativas = 1000
acertos = 0

for _ in range(tentativas):
    if random.random() > 0.5:
        acertos += 1

print('Probabilidade aproximada:', acertos/tentativas)
```
